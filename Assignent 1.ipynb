# Import section

```python
import time
import matplotlib.pyplot as plt
```

# [Problem 1] Create using the exponentiation arithmetic operator 

```python
"""
Code to calculate the thickness when the paper is folded once
"""
THICKNESS = 0.00008
folded_thickness = THICKNESS*2**43
print("Thickness: {} meters".format(folded_thickness))
```

# [Problem 2] Unit Conversion

```python
# Convert meters to kilometers and display with two decimal places
print("Thickness: {: .2f} thousand kilometers".format(folded_thickness/10000000))
```

# [Problem 3] Create using a for statement

```python
"""
Code to calculate the thickness when the paper is folded once
"""
folded_thickness_statement = THICKNESS*1024*1024*1024*1024*2*2*2 # 1024 = 2^10 so 1024*1024 = 2^10*2^10 = 2^20 and so on
print("Thickness: {} meters".format(folded_thickness_statement))
```

# [Problem 4] Comparison of calculation time

```python
%%timeit
start = time.time()
#####
folded_thickness = THICKNESS*2**43
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
%%timeit
start = time.time()
#####
folded_thickness_statement = THICKNESS*1024*1024*1024*1024*2*2*2
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

# Conculse 

Using exponential to calculate the thickness of the sheet is a little faster than using only "*" arithmetic notation 

# [Problem 5] Saving to a list

```python
thickness_list = []
current_thickness = THICKNESS
for i in range(44):
    thickness_list.append(current_thickness)
    current_thickness = current_thickness *2
print("Element of the list:\n",thickness_list)
print ("Length of the list ",len(thickness_list))
```

# [Problem 6] Displaying a line graph

```python
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list) # Enter the variable name of the list in "List name"
plt.show()
```

# Analyze 

The thickness of the paper increased exponentialy after each fold. It mean that paper's thickness will increase slowly in the few first folds and it will increase dramaticaly in some later folds

# [Problem 7] Customizing graphs

```python
"""
Display a red line graph.
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.grid()
plt.plot(thickness_list,".:b",lw=3,markersize=15,markerfacecolor="red") # Enter the variable name of the list in "List name"
plt.show()
```
